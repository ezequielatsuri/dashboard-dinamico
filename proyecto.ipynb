```python
def generar_clave(folio):
    folio_str = str(folio)
    if len(folio_str) == 9:
        return int("0" + folio_str[0])  # Agregar un 0 antes del primer dígito y convertir a entero
    elif len(folio_str) == 10:
        return int(folio_str[:2])  # Tomar los primeros dos dígitos y convertir a entero
    else:
        return None  # Manejo de caso de longitud inesperada

# Aplicar la función y generar la nueva columna 'claveEntidad'
gastos2018['claveEntidad'] = gastos2018['folioviv'].apply(generar_clave)
gastos2020['claveEntidad'] = gastos2020['folioviv'].apply(generar_clave)
gastos2022['claveEntidad'] = gastos2022['folioviv'].apply(generar_clave)
```

```python
import pandas as pd

# Cargar los datos de tus dataframes (sustituye por la forma en que cargas tus datos)
# Aquí simplemente se supone que ya tienes `gastos2018`, `gastos2020`, y `gastos2022`

# Tablas de referencia
lugar_comp_mapping = {
    0: "No aplica", 1: "Mercado", 2: "Tianguis o mercado sobre ruedas", 3: "Vendedores ambulantes",
    4: "Tiendas de abarrotes", 5: "Tiendas específicas del hogar", 6: "Supermercados",
    7: "Tiendas departamentales", 8: "Compras fuera del país", 9: "Tiendas con membresía",
    10: "Tiendas de conveniencia", 11: "Restaurantes", 12: "Loncherías, fondas, torterías",
    13: "Cafeterías", 14: "Pulquería, cantina o bar", 15: "Diconsa", 16: "Lechería Liconsa",
    17: "Persona particular", 18: "Internet"
}

forma_pag_mapping = {
    0: "No aplica", 1: "Efectivo", 2: "Fiado (persona)", 3: "Domiciliación",
    4: "Transferencia", 5: "Tarjeta de crédito", 6: "Tarjeta de débito",
    7: "Cheque", 8: "Vale", 9: "Pago móvil", 10: "Otro"
}

# Reemplazar en los dataframes de gastos
for gastos in [gastos2018, gastos2020, gastos2022]:
    gastos["lugar_comp"] = gastos["lugar_comp"].replace(lugar_comp_mapping)
    gastos["forma_pag1"] = gastos["forma_pag1"].replace(forma_pag_mapping)

# Verifica los cambios
print(gastos2018.head())
print(gastos2020.head())
print(gastos2022.head())
```

```python
# Concatenar los DataFrames en un solo archivo
datos_concatenados = pd.concat([gastos2018, gastos2020, gastos2022], ignore_index=True)


dataPoblacion = pd.read_excel('dataPoblacion.xlsx')
dataCategoriGasto = pd.read_csv('categorias.csv')
dfpGastos = pd.merge(datos_concatenados,dataPoblacion,on='claveEntidad')
dfpcategorias = pd.merge(dfpGastos,dataCategoriGasto,on='clave')
#dfpPlanea.to_csv('datos_unificados_poblacion.csv', index=False, encoding='utf-8')
dfpcategorias.to_csv('gastosUnificados.csv', index=False, encoding='utf-8')

# Exportar el DataFrame combinado a un archivo CSV
#datos_concatenados.to_csv('datos_unificados.csv', index=False, encoding='utf-8')

print("Archivo unificado creado con éxito: 'datos_unificados_1.xlsx'")
```
